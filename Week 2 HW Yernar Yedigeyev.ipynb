```python
usarrests_df = spark.read.csv(
    "/Volumes/databricks_rdatasets_compilation/vf58caff/csv/datasets/USArrests.csv",
    header=True,
    inferSchema=True
)
```

```python
for col_name in usarrests_df.columns:
    usarrests_df = usarrests_df.withColumnRenamed(
        col_name,
        col_name.replace(".", "_")
    )
```

```python
usarrests_df.write \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("usarrests")
```

```python
%sql
SELECT
    ROUND(AVG(Murder), 2) AS avg_murder,
    ROUND(AVG(Assault), 2) AS avg_assault,
    ROUND(AVG(Rape), 2) AS avg_rape
FROM usarrests;
```
