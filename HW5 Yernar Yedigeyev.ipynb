```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import pandas as pd
```

```python
diamonds = sns.load_dataset("diamonds")
diamonds.head()
```

```python
avg_price = diamonds.groupby("cut")["price"].mean().reset_index()
avg_price
```

```python
plt.figure(figsize=(8,5))

plt.bar(avg_price["cut"], avg_price["price"], color="steelblue")

plt.xlabel("Cut")
plt.ylabel("Average Price")
plt.title("Average Diamond Price by Cut")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(x="cut", y="price", data=diamonds)

plt.title("Diamond Price Distribution by Cut")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.violinplot(x="cut", y="price", data=diamonds)

plt.title("Diamond Price Distribution Shape by Cut")

plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12,10))
```

```python
carrier_counts = flights_pd["carrier"].value_counts()

axes[0,0].bar(carrier_counts.index, carrier_counts.values)

axes[0,0].set_title("Flights per Carrier")
axes[0,0].set_xlabel("Carrier")
axes[0,0].set_ylabel("Number of Flights")
```

```python
delay_by_month = flights_pd.groupby("month")["dep_delay"].mean()

axes[0,1].plot(delay_by_month.index, delay_by_month.values, marker="o")

axes[0,1].set_title("Average Departure Delay by Month")
axes[0,1].set_xlabel("Month")
axes[0,1].set_ylabel("Average Delay (minutes)")
```

```python
filtered = flights_pd[(flights_pd["dep_delay"] >= -60) & (flights_pd["dep_delay"] <= 120)]

axes[1,0].hist(filtered["dep_delay"], bins=40, alpha=0.7)

axes[1,0].set_title("Departure Delay Distribution")
axes[1,0].set_xlabel("Delay (minutes)")
```

```python
top5 = flights_pd["carrier"].value_counts().head(5).index
top_data = flights_pd[flights_pd["carrier"].isin(top5)]

sns.boxplot(x="carrier", y="arr_delay", data=top_data, ax=axes[1,1])

axes[1,1].set_title("Arrival Delay by Carrier")
```
