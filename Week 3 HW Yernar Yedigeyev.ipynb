```python
%sql
SELECT *
FROM diamonds
WHERE cut = 'Ideal'
  AND carat > 2
ORDER BY carat DESC
LIMIT 10;
```

```python
%sql
SELECT
    year, month, day, carrier, flight, dep_delay, arr_delay
FROM flights
WHERE month = 12
  AND dep_delay > 120
ORDER BY dep_delay DESC
LIMIT 10;
```

```python
%sql
SELECT
    cut,
    color,
    COUNT(*) AS count,
    ROUND(AVG(price), 2) AS avg_price
FROM diamonds
GROUP BY cut, color
ORDER BY avg_price DESC;
```

```python
%sql
SELECT
    month,
    COUNT(*) AS num_flights,
    ROUND(AVG(arr_delay), 2) AS avg_arrival_delay
FROM flights
WHERE arr_delay IS NOT NULL
GROUP BY month
ORDER BY avg_arrival_delay DESC;
```

```python
%sql
SELECT
    a.name AS airport_name,
    COUNT(*) AS num_flights,
    ROUND(AVG(f.arr_delay), 2) AS avg_arrival_delay
FROM flights f
JOIN airports a
    ON f.dest = a.faa
WHERE f.arr_delay IS NOT NULL
GROUP BY a.name
HAVING COUNT(*) > 100
ORDER BY avg_arrival_delay DESC
LIMIT 10;
```
